```python
import base64
```

```python
#import images from /images directory and convert to base64
def image_to_base64(image_path):
    with open(image_path, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    return encoded_string
# List of image paths
image_paths = [
    "images/Crossover-and-mutation.png",
    "images/darwin-day-2019-i-principali-errori-sulla-teoria-dellevoluzione-che-abbiamo-imparato-a-scuola.jpg",
    "images/darwin4.webp",
    "images/ImageForArticle_25252_17258789538048577.webp"    
]
# Create a dictionary to hold the base64 strings
base64_images = {}
# Loop through the image paths and convert each to base64
for image_path in image_paths:
    # Get the image name without the path
    image_name = image_path.split("/")[-1]
    # Convert to base64 and store in the dictionary
    base64_images[image_name] = image_to_base64(image_path)
# Print the base64 strings and save them to a file divided by comma, without names
file_name = 'dataset2.txt'
for image_name, base64_string in base64_images.items():
    print(base64_string)
    with open(file_name, 'a') as f:
        f.write(base64_string + ',')

    

```

```python
#given a base64 string, decode it and print the image in the notebook
base64_string = base64_images['darwin4.webp']
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib.image as mpimg
import io
import PIL.Image as Image

# Decode the base64 string
image_data = base64.b64decode(base64_string)
# Convert the byte data to a numpy array
image_array = np.frombuffer(image_data, dtype=np.uint8)
# Decode the image array using OpenCV
image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)
# Convert the image from BGR to RGB format
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Display the image using matplotlib
plt.imshow(image_rgb)




```

```python
#add random mutation to the string and print again the image
import random
import string
def random_mutation(base64_string, mutation_rate=0.01):
    # Convert the base64 string to a byte array
    byte_array = bytearray(base64.b64decode(base64_string))
    
    # Apply random mutations
    for i in range(len(byte_array)):
        if random.random() < mutation_rate:
            # Randomly change the byte
            byte_array[i] = random.randint(0, 255)
    
    # Convert the mutated byte array back to a base64 string
    mutated_base64_string = base64.b64encode(byte_array).decode('utf-8')
    
    return mutated_base64_string
# Apply random mutation to the base64 string
mutated_base64_string = random_mutation(base64_string, mutation_rate=0.01)
# Decode the mutated base64 string
mutated_image_data = base64.b64decode(mutated_base64_string)
# Convert the byte data to a numpy array
mutated_image_array = np.frombuffer(mutated_image_data, dtype=np.uint8)
# Decode the image array using OpenCV
mutated_image = cv2.imdecode(mutated_image_array, cv2.IMREAD_COLOR)
# Convert the image from BGR to RGB format
mutated_image_rgb = cv2.cvtColor(mutated_image, cv2.COLOR_BGR2RGB)
# Display the mutated image using matplotlib
plt.imshow(mutated_image_rgb)
```

```python
csv_file = '/home/lucapernice/NLP_project/logs/b5a073fb-e7af-4243-a517-51989e91bf8a.csv'
import pandas as pd

# Read the CSV file
df = pd.read_csv(csv_file)
```

```python
valid_df = df[df['integrity_check'] == True]
```

```python
valid_df
```

```python
#average fitness for generation
avg_fitness = valid_df.groupby('generation')['fitness'].mean().reset_index()
best_fitness = valid_df.groupby('generation')['fitness'].max().reset_index()
best_compression = valid_df.groupby('generation')['compression_ratio'].max().reset_index()
# Plotting the average and best fitness over generations
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 5))
plt.plot(avg_fitness['generation'], avg_fitness['fitness'], label='Average Fitness', color='blue')
plt.plot(best_fitness['generation'], best_fitness['fitness'], label='Best Fitness', color='red')
#write the best fitness as a label over the best fitness
for i, row in best_fitness.iterrows():
    plt.text(row['generation'], row['fitness'], f"{row['fitness']:.2f}", fontsize=10, ha='center', va='bottom', bbox=dict(facecolor='white', alpha=0.5, pad=0.1))


plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Average and Best Fitness Over Generations')
plt.legend()
plt.grid()
plt.show()

```

```python
# Plotting the best compression ratio over generations
plt.figure(figsize=(20, 5))
plt.plot(best_compression['generation'], best_compression['compression_ratio'], label='Best Compression Ratio', color='green')
#write the best compression as a label over the best compression
for i, row in best_compression.iterrows():
    plt.text(row['generation'], row['compression_ratio'], f"{row['compression_ratio']:.6f}", fontsize=10, ha='center', va='bottom', bbox=dict(facecolor='white', alpha=0.5, pad=0.1))
plt.xlabel('Generation')
plt.ylabel('Compression Ratio')
plt.title('Best Compression Ratio Over Generations')
plt.legend()
plt.grid()
plt.show()
```
